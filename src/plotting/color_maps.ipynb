```python
import matplotlib.pyplot as plt
import seaborn as sns
import anndata as ad
import glasbey
import tol_colors
from matplotlib.colors import ListedColormap, to_rgb
import json
```

# Method colors

```python
colors = sns.palettes.color_palette("husl", 5)
methods = ["Supervised", "Unsupervised", "Prior-Knowledge-driven", 'Pre-trained Models', "Baselines"]
```

```python
color_mapping_methods = {
    method: to_rgb(col)
    for method, col in zip(methods, colors)
}
```

```python
method_colors = ListedColormap([color_mapping_methods[method] for method in methods])
```

```python
method_colors
```

```python
with open("method_colors.json", "w") as f:
    json.dump(color_mapping_methods, f)
```

# Dataset colors

```python
palette = glasbey.create_palette(palette_size=9, colorblind_safe=True)
datasets = ['cHL_2_MIBI', 'IMMUcan']
```

```python
color_mapping_datasets = {
    dataset: to_rgb(palette) for dataset, palette in zip(datasets, palette)
}
```

```python
color_mapping_datasets
```

```python
dataset_colors = ListedColormap([color_mapping_datasets[dataset] for dataset in datasets])
dataset_colors
```

```python
with open("dataset_colors.json", "w") as f:
    json.dump(color_mapping_datasets, f)
```
