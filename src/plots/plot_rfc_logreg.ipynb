```python
data = []
```

```python
datasets = [AML_results, healthy_BM_results, BE_tonsil_results, tonsil_training_results, CRC_FFPE_results, feto_maternal_results, intestine_CODEX_results, lymphoma_CODEX_results, cHL1_MIBI_results, cHL2_MIBI_results, cHL_CODEX_results, MRL_CODEX_results, sarc_results, TB_results, tonsil_CODEX_results, tonsil_CODEX2_results]
```

```python
dataset_names = ['AML_results', 'healthy_BM_results', 'BE_tonsil_results', 'tonsil_training_results', 'CRC_FFPE_results', 'feto_maternal_results', 'intestine_CODEX_results', 'lymphoma_CODEX_results', 'cHL1_MIBI_results', 'cHL2_MIBI_results', 'cHL_CODEX_results', 'MRL_CODEX_results', 'sarc_results', 'TB_results', 'tonsil_CODEX_results', 'tonsil_CODEX2_results']
```

```python
for dataset, dataset_name in zip(datasets, dataset_names):
    for classifier_name, metrics in dataset.items():
        for metric_name, metric_value in metrics.items():
            data.append({
                'Dataset': dataset_name,
                'Classifier': classifier_name,
                'Metric': metric_name,
                'Value': metric_value
            })
```

```python
print(data)
```

```python
plt.style.use('ggplot')
```

```python
df = pd.DataFrame(data)
accuracy_df = df[df['Metric'] == 'average_accuracy']
plt.figure(figsize=(16, 8))
sns.barplot(x='Dataset', y='Value', hue='Classifier', data=accuracy_df, width=0.5)

# Customize the plot
plt.title('Accuracy Comparison: RFC vs LogReg across Datasets', fontsize=16)
plt.xlabel('Dataset', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.legend(title='Classifier', title_fontsize='12', fontsize='10')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)
accuracy_df = df[df['Metric'] == 'average_weighted_f1_score']
plt.figure(figsize=(16, 8))
sns.barplot(x='Dataset', y='Value', hue='Classifier', data=accuracy_df, width=0.5)

# Customize the plot
plt.title('Weighted F1-score: RFC vs LogReg across Datasets', fontsize=16)
plt.xlabel('Dataset', fontsize=12)
plt.ylabel('weighted f1', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.legend(title='Classifier', title_fontsize='12', fontsize='10')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
