```python

from models_classes.gridsearch import Tune_Eval
import pandas as pd
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
```

```python
rs = 20240925
param_grid = {
    'n_estimators': [100, 250],
    'max_depth': [20, 25, 30, 40],
    'min_samples_split' : [2, 3],
    'min_samples_leaf' : [1, 2]

}
```

```python
df1 = pd.read_csv('/Volumes/Lukas_SSD/phenotyping_benchmark/datasets/breast_fibro_IMC/quantification/processed/kfolds/fold_1_test.csv')
df2 = pd.read_csv('/Volumes/Lukas_SSD/phenotyping_benchmark/datasets/breast_fibro_IMC/quantification/processed/kfolds/fold_1_train.csv')
df = pd.concat([df1, df2])
y = df['encoded_phenotype']
class_weights = compute_class_weight('balanced', classes=np.unique(y), y=y)
class_weight_dict = {cls: weight for cls, weight in zip(np.unique(y), class_weights)}
breast_fibro_IMC = Tune_Eval(class_weight=class_weight_dict, n_jobs=6, random_state=rs, model='random_forest')
breast_fibro_IMC.train_tune_evaluate(path='/Volumes/Lukas_SSD/phenotyping_benchmark/datasets/AML_bone_marrow/quantification/healthy_BM/processed/kfolds', param_grid = param_grid,)
breast_fibro_IMC.save_results(save_path='/Volumes/Lukas_SSD/phenotyping_benchmark/datasets/breast_fibro_IMC/results_rfc/', label_path='/Volumes/Lukas_SSD/phenotyping_benchmark/datasets/AML_bone_marrow/quantification/healthy_BM/processed/labels.csv')
```
