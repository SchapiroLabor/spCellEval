```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
result_data = pd.read_csv('/Users/lukashat/Documents/PhD_Schapiro/Projects/phenotype_benchmark/results/cHL_2_MIBI/TACIT/level3/predictions_1.csv')
```

```python
cr = classification_report(result_data['true_phenotype'], result_data['predicted_phenotype'])
print(cr)
```

```python
cm = confusion_matrix(result_data['true_phenotype'], result_data['predicted_phenotype'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(result_data['true_phenotype']))
disp.plot(cmap='Blues', xticks_rotation=90)
```
