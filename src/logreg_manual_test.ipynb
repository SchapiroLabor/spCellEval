```python
import os
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, f1_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv('/Volumes/Lukas_SSD/phenotyping_benchmark/datasets/Maps_data/cHL_1_MIBI/quantification/processed/kfolds/fold_1_train.csv')
test = pd.read_csv('/Volumes/Lukas_SSD/phenotyping_benchmark/datasets/Maps_data/cHL_1_MIBI/quantification/processed/kfolds/fold_1_test.csv')
X_train = train.drop(columns=['encoded_phenotype'])
y_train = train['encoded_phenotype']
X_test = test.drop(columns=['encoded_phenotype'])
y_test = test['encoded_phenotype']
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000, n_jobs=-1, multi_class='multinomial')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
model = LogisticRegression(max_iter=1000, n_jobs=-1, multi_class='ovr')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
